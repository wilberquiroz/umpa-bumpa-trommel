```python
def scale_range(v, y):
    """
    Escala un vector manteniendo el primer elemento fijo, de manera que
    el Ãºltimo menos el primero pase a valer (x + y), donde x es el
    rango original.

    v: lista [v0, v1, ..., vn]
    y: incremento deseado del rango

    return: nueva lista escalada
    """
    if len(v) < 2:
        return v[:]  # no tiene rango

    v0 = v[0]
    x = v[-1] - v0  # rango original

    if x == 0:
        # No se puede escalar un rango cero, extender todos por y
        return [v0 + y for _ in v]

    factor = (x + y) / x

    return [v0 + (e - v0) * factor for e in v]
```

```python
scale_range(v, 0.9)
```
